#   Data Prep 
    1.  Rename column names and make them all lowercase
    2.  Data Conversion 
        a.  Year_of_Release to Int64

    3.  Why datatypes chaged and data description

    4.  Missing values

        b.  user_score missing values could be augmented from mean critic_score plus adjustment, or the mean critic_score of games in the same genre and yr, or leave as NaN, or just dropped. TBD is a common placeholder here, what dios it mean?

        c. are rating missing values important? if they are missing we copuld augment by assigning the same top rating for games in the same genre.

    5.  Fill missing values / leave NaN
        a.  critic_score missing values will be augmented from NaN by the user_score where available with an adjustment, or the mean of that genre and year will be used, if still NaN the data will be dropped.

        b. user_score TBD will be treated the same as NaN. Missing values will be replaced by the critic_score values where available plus an adjustment, replaced by the mean user_score rating for that game genre and year, or droped if nothing else is available. 

        c. Rating missing values will be assigned the top value of the games in the same genere / yr.

    6. Why are values missing?
        Missing values are a sign of missing collection data. On some data values might simply not exist. Here are my thoughts for each specific column:
        a.  critic_score missing values could mean that no critic has reviewed the game.
        b.  user_score simply does not exist because it's too new (TBD), or simply statistics have not been collected where NaN.
        c.  Rating missing values could be because the game is not ESRB certified.
    
    7. TBD values in user_score
        This only pops up in the user_score column. the approach will follow like this:
        a.  TBD and NaN will be treated the same
        b.  NaN values will be replaced with critic_score values plus some adjustment
        c.  remaining rows with NaN in this column will be replaced with the mean of the game's genere/yr
        d.  remaining NaN values will be dropped.

    8. Scale user_score to a scale of 100, or just multiply by 10. I'd prefer this to be on the same scale as critic_score

```python
def data_prep(df):
    # 1. Rename column names and make them all lowercase
    df.columns = [col.lower() for col in df.columns]
    
    # 2. Data Conversion
    # a. Year_of_Release to Int64
    df['year_of_release'] = df['year_of_release'].astype('Int64')
    
    # b. User_Score to float64 and scale
    def convert_user_score(value):
        try:
            return float(value)*10
        except ValueError:
            return np.nan
        
    df['user_score'] = df['user_score'].apply(convert_user_score)
    
    # Calculate the mean difference between critic_score and user_score
    valid_scores = df.dropna(subset=['critic_score', 'user_score'])
    critic_to_user_adjustment = (valid_scores['critic_score'] - valid_scores['user_score']).mean()
    user_to_critic_adjustment = (valid_scores['user_score'] - valid_scores['critic_score']).mean()
    
    # 4. Missing values handling
    # a. Augment critic_score from user_score plus an adjustment or use mean critic_score of games in the same genre and year
    def fill_critic_score(row):
        if pd.isna(row['critic_score']):
            if not pd.isna(row['user_score']):
                if not (row['user_score'] + critic_to_user_adjustment > 100):  # Use the calculated adjustment
                    return  row['user_score'] + critic_to_user_adjustment
                else: return 100
            else:
                genre_year_mean = df[(df['genre'] == row['genre']) & (df['year_of_release'] == row['year_of_release'])]['critic_score'].mean()
                return genre_year_mean
        return row['critic_score']
    
    df['critic_score'] = df.apply(fill_critic_score, axis=1)
    
    # b. Augment user_score from critic_score plus adjustment or use mean user_score of games in the same genre and year
    def fill_user_score(row):
        if pd.isna(row['user_score']):
            if not pd.isna(row['critic_score']):
                if not (row['critic_score'] + user_to_critic_adjustment > 100): # Use the calculated adjustment
                    return row['critic_score'] + user_to_critic_adjustment
                else: return 100
            else:
                genre_year_mean = df[(df['genre'] == row['genre']) & (df['year_of_release'] == row['year_of_release'])]['user_score'].mean()
                return genre_year_mean
        return row['user_score']
    
    df['user_score'] = df.apply(fill_user_score, axis=1)
    
    # c. Assign the top rating for games in the same genre/year if Rating is missing
    def fill_rating(row):
        if pd.isna(row['rating']):
            mode_rating = df[(df['genre'] == row['genre']) & (df['year_of_release'] == row['year_of_release'])]['rating'].mode()
            if not mode_rating.empty:
                return mode_rating.iloc[0]
        return row['rating']
    
    df['rating'] = df.apply(fill_rating, axis=1)
    
    # 5. Replace remaining NaN values
    def fill_critic_score_with_genre_year_mean(row):
        if pd.isna(row['critic_score']):
            genre_year_mean = df[(df['genre'] == row['genre']) & (df['year_of_release'] == row['year_of_release'])]['critic_score'].mean()
            return genre_year_mean
        return row['critic_score']
    
    def fill_user_score_with_genre_year_mean(row):
        if pd.isna(row['user_score']):
            genre_year_mean = df[(df['genre'] == row['genre']) & (df['year_of_release'] == row['year_of_release'])]['user_score'].mean()
            return genre_year_mean
        return row['user_score']
    
    def fill_rating_with_genre_year_mode(row):
        if pd.isna(row['rating']):
            mode_rating = df[(df['genre'] == row['genre']) & (df['year_of_release'] == row['year_of_release'])]['rating'].mode()
            if not mode_rating.empty:
                return mode_rating.iloc[0]
        return row['rating']
    
    df['critic_score'] = df.apply(fill_critic_score_with_genre_year_mean, axis=1)
    df['user_score'] = df.apply(fill_user_score_with_genre_year_mean, axis=1)
    df['rating'] = df.apply(fill_rating_with_genre_year_mode, axis=1)
    
    # Drop rows with any remaining NaN values
    df.dropna(subset=['critic_score', 'user_score', 'rating', 'year_of_release'], inplace=True)
    
    # Calculate total_sales
    df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
 
    return df

df = data_prep(df)
df.info()
```

#   Approach

I'm not a magician, nor do i have a crystal ball. For proper predictions on time series data I would use an RNN, however it would sufice here to leverage trends lines and fit a curve for each platform. The curve fit function could be linear (y=mx+b), or polynomial. The data for all platform sales (except PC) lookls like a hyperbole, therefore I'd rather opt for polyniomial regression.

#   Steps

1.  Isolate the last 11 years of sales of each platform.
2.  Calculate trendlines using polynomial regression.
3.  Use the polynomial to generate a prediction dataframe for each console sales for 2017. x=2017
4.  Combine predicted data frame and previous 11 years of data and graph.

</b></div>


```python
# Function to predict sales for the year 2017 using polynomial regression
def predict_sales(df, top_platforms, years_to_consider, prediction_year):
    predictions = {}
    for platform in top_platforms:
        platform_data = df[df['platform'] == platform]
        recent_data = platform_data[platform_data['year_of_release'] >= (platform_data['year_of_release'].max() - years_to_consider)]
        x = recent_data['year_of_release'].values
        y = recent_data['total_sales'].values
        
        if len(x) > 0 and len(y) > 0:
            # Ensure x and y are numeric
            x = x.astype(int)
            y = y.astype(float)
            
            coeffs = np.polyfit(x, y, 2)  # Fit a 2nd degree polynomial
            poly = np.poly1d(coeffs)
            
            y_pred = poly(prediction_year)
            
            predictions[platform] = y_pred
    
    return predictions

# Aggregate total sales by platform
platform_sales = df.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
top_platforms = platform_sales.head(10).index.tolist()

# Predict sales for the year 2017 using previous 8 years of data
predictions = predict_sales(df, top_platforms,8,2017)

# Combine the predicted data 
prediction_df = pd.DataFrame(predictions.items(), columns=['platform', 'total_sales'])
prediction_df['year_of_release'] = 2017

# Filter the original dataframe to include only the last 8 years of data
max_year = df['year_of_release'].max()
filtered_df = df[df['year_of_release'] >= (max_year - 8)]

# Combine filtered data with predictions
combined_df = pd.concat([filtered_df, prediction_df])

# Plot the combined data
fig = px.bar(combined_df, x='year_of_release', y='total_sales', color='platform', title='Total Sales Distribution by Year for Top Platforms',
             labels={'year_of_release': 'Year of Release', 'total_sales': 'Total Sales (in millions)'}, barmode='group')

fig.show()
```

# 7. Take a look at how user and professional reviews affect sales for one popular platform (you choose). Build a scatter plot and calculate the correlation between reviews and sales. Draw conclusions.
 </b></div>

```python
# Function to analyze the impact of reviews on sales for each platform
def analyze_reviews_impact(df,platforms=[]):
    
    if platforms == []:
        platforms = df['platform'].unique()
    results = []

    for platform in platforms:
        platform_df = df[df['platform'] == platform]
        
        correlation_critic_sales = platform_df['critic_score'].corr(platform_df['total_sales'])
        correlation_user_sales = platform_df['user_score'].corr(platform_df['total_sales'])
        
        results.append({
            'platform': platform,
            'correlation_critic_sales': correlation_critic_sales,
            'correlation_user_sales': correlation_user_sales
        })
        
        fig_critic = px.scatter(platform_df, x='critic_score', y='total_sales', title=f'Critic Scores vs. Sales for {platform}', labels={'critic_score': 'Critic Score', 'total_sales': 'Total Sales (in millions)'})
        fig_user = px.scatter(platform_df, x='user_score', y='total_sales', title=f'User Scores vs. Sales for {platform}', labels={'user_score': 'User Score', 'total_sales': 'Total Sales (in millions)'})
        
        fig_critic.show()
        fig_user.show()
    
    return pd.DataFrame(results)



#Analyze correlation and assign label
def correlation_lvl(df):
    levels = {'Strong': 0.7,
              'Moderate': 0.5,
              'Weak': 0.3,
              'Negligible': 0}

    # Initialize new columns
    df['correlation_critic_sales_lvl'] = 'Unknown'
    df['correlation_user_sales_lvl'] = 'Unknown'

    for index, row in df.iterrows():
        if pd.isna(row['correlation_critic_sales']):
            df.loc[index, 'correlation_critic_sales_lvl'] = 'Unknown'
        elif abs(row['correlation_critic_sales']) >= levels['Strong']:
            df.loc[index, 'correlation_critic_sales_lvl'] = 'Strong'
        elif abs(row['correlation_critic_sales']) >= levels['Moderate']:
            df.loc[index, 'correlation_critic_sales_lvl'] = 'Moderate'
        elif abs(row['correlation_critic_sales']) >= levels['Weak']:
            df.loc[index, 'correlation_critic_sales_lvl'] = 'Weak'
        else:
            df.loc[index, 'correlation_critic_sales_lvl'] = 'Negligible'

        if pd.isna(row['correlation_user_sales']):
            df.loc[index, 'correlation_user_sales_lvl'] = 'Unknown'
        elif abs(row['correlation_user_sales']) >= levels['Strong']:
            df.loc[index, 'correlation_user_sales_lvl'] = 'Strong'
        elif abs(row['correlation_user_sales']) >= levels['Moderate']:
            df.loc[index, 'correlation_user_sales_lvl'] = 'Moderate'
        elif abs(row['correlation_user_sales']) >= levels['Weak']:
            df.loc[index, 'correlation_user_sales_lvl'] = 'Weak'
        else:
            df.loc[index, 'correlation_user_sales_lvl'] = 'Negligible'
    
    return df

# Analyze the impact of reviews on sales for each platform
# Pass an empty list and it will generate graphs for all platforms
results_df = analyze_reviews_impact(df,['Wii','PS'])
print(results_df)
# Assign level label
results_df = correlation_lvl(results_df)
print(results_df)
```
